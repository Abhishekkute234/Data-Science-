# MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np

# Create simple data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic line plot
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Multiple lines with different styles
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='--')
plt.plot(x, y2, label='cos(x)', color='red', linewidth=2)
plt.title("Trigonometric Functions")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create random data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar(label='Color Intensity')
plt.title("Bubble Chart")
plt.xlabel("X Variable")
plt.ylabel("Y Variable")
plt.show()
```

```python
# Bar chart example
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color=['red', 'green', 'blue', 'cyan'])
plt.title("Category Values")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Histogram with normal distribution
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1
axes[0,0].plot(x, y1, color='blue')
axes[0,0].set_title("Sine Wave")

# Plot 2
axes[0,1].scatter(x, y2, color='red')
axes[0,1].set_title("Cosine Points")

# Plot 3
axes[1,0].bar(categories, values)
axes[1,0].set_title("Bar Chart")

# Plot 4
axes[1,1].hist(data, bins=30)
axes[1,1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

```python
# Advanced customization
plt.figure(figsize=(10, 6))

x = np.linspace(0, 10, 100)
y = np.exp(x)

plt.plot(x, y, 
         linewidth=2, 
         linestyle=':', 
         color='green',
         marker='o',
         markersize=4,
         markerfacecolor='red',
         markeredgecolor='black')

plt.title("Exponential Growth", fontsize=14, fontweight='bold')
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.yscale('log')
plt.show()
```

```python
# Pie chart example
labels = ['Python', 'C++', 'Java', 'JavaScript']
sizes = [45, 30, 15, 10]
explode = (0.1, 0, 0, 0)

plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        autopct='%1.1f%%',
        shadow=True,
        startangle=90)
plt.axis('equal')
plt.title("Programming Language Popularity")
plt.show()
```

```python
# Heatmap example
data = np.random.rand(10, 12)
plt.imshow(data, cmap='coolwarm')
plt.colorbar()
plt.title("Heatmap Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 3D plotting (requires mpl_toolkits)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

ax.scatter(x, y, z, c=z, cmap='viridis')
ax.set_title("3D Scatter Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
# Available styles
print(plt.style.available)

# Using predefined style
plt.style.use('ggplot')

x = np.linspace(0, 10)
y = np.random.randn(50).cumsum()

plt.plot(x, y)
plt.title("Styled Plot with ggplot")
plt.show()
```

```python
# Box plot example
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, 
            vert=True, 
            patch_artist=True,
            labels=['Set 1', 'Set 2', 'Set 3'])
plt.title("Box Plot Example")
plt.show()
```

```python
# Error bar example
x = np.arange(0, 10)
y = 2.5 * np.sin(x / 20 * np.pi)
error = 0.2 + 0.2 * x

plt.errorbar(x, y, yerr=error, fmt='-o')
plt.title("Measurement with Error Bars")
plt.show()
```
