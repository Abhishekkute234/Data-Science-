# SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Set Seaborn style
sns.set_theme(style="whitegrid")
print("Seaborn version:", sns.__version__)
```

```python
# Load sample dataset
tips = sns.load_dataset("tips")
print("Tips Dataset Head:")
print(tips.head())

# Line plot
plt.figure(figsize=(8, 4))
sns.lineplot(x="day", y="total_bill", data=tips)
plt.title("Daily Total Bills")
plt.show()
```

```python
# Scatter plot with hue
sns.scatterplot(x="total_bill", y="tip", 
                hue="time", style="sex",
                data=tips, palette="viridis")
plt.title("Tips vs Total Bill")
plt.show()
```

```python
# Histogram with KDE
sns.histplot(data=tips, x="total_bill", kde=True, bins=20)
plt.title("Distribution of Total Bills")
plt.show()
```

```python
# Bar plot
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("Average Bill by Day and Gender")
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 5))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips)
plt.title("Box Plot of Bills by Day and Smoking Status")
plt.show()

# Violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Daily Bills")
plt.show()
```

```python
# Pair plot
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species", palette="Set2")
plt.suptitle("Pair Plot of Iris Dataset")
plt.show()
```

```python
# Regression plot
sns.lmplot(x="total_bill", y="tip", data=tips, 
           hue="smoker", markers=["o", "x"])
plt.title("Tip vs Total Bill with Regression Line")
plt.show()
```

```python
# Facet grid
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
g.fig.suptitle("Facet Grid of Tips Data")
plt.tight_layout()
plt.show()
```

```python
# Show color palette
print("Current palette:", sns.color_palette("husl"))

# Use custom palette
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set3")
plt.title("Custom Color Palette Example")
plt.show()
```

```python
# Create subplot with matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips, ax=ax)
ax.set_title("Customized Plot")
ax.set_xlabel("Total Bill ($)")
ax.set_ylabel("Tip ($)")
ax.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
# Correlation heatmap
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
